# 01 Extraction



Objective: To ingest the raw dataset from the source and perform an initial audit of the data quality.

Data Source: Kaggle - Credit/Loan Dataset - Rural India.
https://www.kaggle.com/datasets/heydido/creditloan-dataset-rural-india?resource=download 

Extraction Method: CSV File Ingestion.

Raw File Location: data/raw/RuralCreditData.csv

```python
import pandas as pd

# 1. Define the source of the data
# Even though the file is already on your computer, 
# we write this code to show the "ingestion" process.
file_path = '../data/raw/RuralCreditData.csv'

try:
    df_raw = pd.read_csv(file_path)
    print("Data Extraction Successful!")
except FileNotFoundError:
    print(" Error: The raw file was not found in data/raw/. Please check the path.")

# Display the first 5 rows to verify extraction
df_raw.head()
```

```python
# 1. Check the dimensions of the raw data
print(f"Raw Dataset Shape: {df_raw.shape}")
print(f"Total Rows: {df_raw.shape[0]}")
print(f"Total Columns: {df_raw.shape[1]}")

# 2. List all columns to identify naming issues (like the misspelled 'water_availabity')
print("\n--- Raw Column Names ---")
print(df_raw.columns.tolist())

# 3. Check for missing values in the raw state
print("\n--- Missing Values Count ---")
print(df_raw.isnull().sum())

# 4. Check data types to see if numbers are being read as strings
print("\n--- Data Types ---")
print(df_raw.dtypes)
```
