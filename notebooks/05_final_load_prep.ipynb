```python
import pandas as pd

# 1. Load the final processed dataset
df_final = pd.read_csv('../data/processed/RuralCreditData_processed.csv')

# 2. Final Sanity Check
print("--- FINAL DATA AUDIT ---")
print(f"Total Rows: {df_final.shape[0]}")
print(f"Total Columns: {df_final.shape[1]}")
print(f"Missing Values: {df_final.isnull().sum().sum()}")

# 3. Verify the Business Metrics are present
required_cols = ['disposable_income', 'loan_to_income_ratio', 'risk_level']
for col in required_cols:
    if col in df_final.columns:
        print(f"✅ {col} is present.")
    else:
        print(f"❌ {col} is missing!")

# 4. Save the final 'Production' file
# We save it as a separate file to distinguish it from the 'processed' one
df_final.to_csv('../data/processed/FINAL_Production_LoanData.csv', index=False)

print("\nFinal Production dataset saved successfully. Ready for Tableau Load.")
```
